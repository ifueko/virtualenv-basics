# Python Virtual Environments for Machine Learning

For certain machine learning applications, dependency conflicts can be a huge headache. In this notebook, we'll go over the basics of creating a virtual environment in python.


One thing to note is that this is unnecessary in a jupyter notebook running with google colab, as the installed system is reset at the start of each runtime. However, we can still create and utilize virtual environments within a colab notebook. When requirements are provided from existing code, this becomes much easier.

Most of our commands will be executed on the command line for the first part of this notebook, using the `! command` notation.

## Installing the package and Creating a Virtual Environment

Vitual environments in python utilize a combination of `pip` for installation and the `virtualenv` command, which (after installation) can also be run using `python -m virtualenv`. 

You can install the virtualenv pacakge using pip:

`pip install virutalenv`

Lets look at a clear example:

````
! pip install virtualenv
````

````
! virtualenv /venv1 # creating a virtual environment using the command baked
                    # into the python module
````

````
! python -m virtualenv /venv2 # creating a virtual environment using the python
                              # module main executionsyntax
````

````
# looking at contents, should be identical.
! ls /venv1
! ls /venv2
````

## Activating the virtual environment and Installing Dependencies

A python virtual environment changes the packages that are installed and available in python. The `pip freeze` command will show you what package are installed. Lets see what packages are installed before we activate our virutal environment:

````
! pip freeze
````

To activate the environment, we use the command `source <path-to-virtualenv>/bin/activate`, in this case using the first environment we made, `source /venv1/bin/activate`. We can then execute `pip freeze` to see what python modules are available:

````
! source /venv1/bin/activate; pip install torch; pip freeze
````

Note that the `! source ...` command in a colab notebook will only preserve our environment for that command state:

````
! source /venv1/bin/activate; pip install torch;
! pip freeze
````

Running `pip freeze` a second time will run our native environment rather than the virtual environment. If we were on the command line, this would be persistent. If we wanted to deactivate our current virtual environment, we would need to call `deactivate`.

`deactivate`

````
! source /venv1/bin/activate; pip freeze | head -5; deactivate; echo 'DEACTIVATED'; pip freeze | head -5;
````

One useful `pip` feature that can be used once in a virtual environment is the


 `pip install -r <requirements-filename>` command. This will install all requirements to the current environment, native or virtual.

Lets say we want to import the open3d module. This is a cool library for 3d data processing. This isn't installed by default, so we could intall this by calling `pip install open3d`. We can also create a `requirements.txt` file with `open3d` as a line and use `pip install -r requirements.txt`:

````
! echo 'open3d' > requirements.txt
! cat requirements.txt

# try to import to make sure its not installed
try:
  import open3d
  print('success')
except:
  print('fail')

# install with pip requirements file
! pip install -r requirements.txt

# try again. should succeed.
try:
  import open3d
  print('success')
except:
  print('fail')
````

This describes the simplest, main flow for utilizing virtual environments in python.
Colab is great because virtually any standard machine learninb library is available with python, and thus no virtual environments are needed.

However, in a local environment this can be useful, as you don't need to install things locally or worry about version constraints. Its also useful when others have provided code, and have used `pip freeze` to store the dependencies in a txt file.


## 2. A Concrete Example: DCGAN

Lets look at a concrete example of forking some code that utilizes a virtual environment:

````
! git clone https://github.com/ifueko/virtualenv-basics.git
! pip install -r virtualenv-basics/dcgan/requirements.txt
! python virtualenv-basics/dcgan/main.py --dataset fashionmnist --cuda
````

After we run, we can see all the images generated and saved model parameters:

````
! ls output
````

### REAL FASHION IMAGES

real_samples.png

### GENERATED (FAKE) FASION IMAGES **EPOCH 24**

fake_samples_epoch_024.png

I like running this with colab because I can change my runtime type to GPU which makes it much faster than on my plain old CPU.
Furthermore in Colab, all of our requirements are already installed. In this example using the Windows subsystem or linux, we can do the following commands:

```
git clone git@github.com:ifueko/virtualenv-basics.git
cd virtualenv-basics
virtualenv venv
source venv/bin/activate
cd dcgan
pip install -r requirements.txt
python main.py --dataset fashionmnist
```

We can see that we are able to run the same code this way using a completely different type of machine with (Windows Subsystem for Linux).
